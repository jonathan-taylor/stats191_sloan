### Election example: best subset

Here is the data for the election example discussed in the notes. We
are going to use two different model selection strategies
to choose a model. The first, *best subset*, tries to find
the best model out of all possible model for a given criterion. In 
*R*, the criterion must be one of $R^2, R^2_a, C_p$.

```python
%load_ext course_modules.knitr_extension
```

```python
%%knitr
url = 'http://stats191.stanford.edu/data/election.table'
election.table = read.table(url, header=T)
pairs(election.table[,2:ncol(election.table)], cex.labels=3, pch=23,
        bg='orange', cex=2)
```

Using $R^2$ doesn't work because the full model (all the variables) 
always wins. The library that does this in *R* is called *leaps*. 

Leaps takes a design matrix as argument: we have throw to away the intercept
column or leaps will complain.


```python
%%knitr
library(leaps)
election.lm = lm(V ~ I + D + W +G:I + P + N, election.table)
X = model.matrix(election.lm)[,-1]
```

We'll plot the three best models for $R^2$ for each model size of model.
As we can see, it is monotone in the size of the model.

```python
%%knitr
election.leaps = leaps(X, election.table$V, nbest=3, method='r2')
plot(election.leaps$size, election.leaps$r2, pch=23, bg='orange', cex=2)
best.model.r2 = election.leaps$which[which((election.leaps$r2 == max(election.leaps$r2))),]
best.model.r2
```

Using adjusted $R^2$ is not monotone

```python
%%knitr
election.leaps = leaps(X, election.table$V, nbest=3, method='adjr2')
plot(election.leaps$size, election.leaps$adjr2, pch=23, bg='orange', cex=2)
best.model.adjr2 = election.leaps$which[which((election.leaps$adjr2 == max(election.leaps$adjr2))),]
best.model.adjr2
```

Nor is Mallow's $C_p$

```python
%%knitr
election.leaps = leaps(X, election.table$V, nbest=3, method='Cp')
plot(election.leaps$size, election.leaps$Cp, pch=23, bg='orange', cex=2)
best.model.Cp = election.leaps$which[which((election.leaps$Cp == min(election.leaps$Cp))),]
best.model.Cp
```

### Election example: stepwise

The next algorithm is an example of a stepwise model building procedure
that tries to include or delete a variable at a time.

The default direction argument is *both*, which tries to add or delete
variables from an initial model.


```python
%%knitr
election.step.both = step(election.lm, direction='both')
summary(election.step.both)
```

An alternative is *backward* which only tries to delete variables.


```python
%%knitr
election.step.backward = step(election.lm, direction='backward')
summary(election.step.backward)
```

Or *forward*, which only tries to add variables to the model. For the forward direction, we must provide *step* with a largest possible model. This is what
the second argument is here. We can also specify a *lower* argument as well.


```python
%%knitr
election.step.forward = step(lm(V ~ 1, election.table), list(upper = ~ I + D + W +G:I + P + N), direction='forward')
summary(election.step.forward)
```

Stepwise search also allows interactions. This next call
searches for all 2-way interaction, only including those 2-way
interactions whose main effects are in the model.


```python
%%knitr
election.step.interactions = step(election.lm, list(upper = ~ .^2), direction='both')
summary(election.step.interactions)
```

To run BIC instead of AIC, we add an extra argument *k*, which is
the multiplier of the number of parameters. Hence, it is 2 for AIC
and *log(n)* for BIC.

```python
%%knitr
election.step.interactions = step(election.lm, list(upper = ~ .^2), direction='both', k=log(21))
summary(election.step.interactions)
```

Let's also confirm our formula for AIC of a multiple linear regression model.

```python
%%knitr
sigma2.mle = sum(resid(election.lm)^2) / 21
21*log(2*pi*sigma2.mle) + 21 + 2 * 8
AIC(election.lm)
```

### Cross-validation

In this example, we will use the list of models produced by *leaps*
and compute a 5-fold cross-validation score for each model.

The cross-validation code is in the *boot* library, which you may have
to install:


```python
%%knitr
library(boot)
V = election.table$V
election.leaps = leaps(X, V, nbest=3, method='Cp')
```

Next, we create a vector to store the cross-validation scores in, and
loop over all models. For subsets of size 1, the *cv.glm* function
seems to behave slightly differently than general subsets, so we treat
it slightly differently.

```python
%%knitr
election.leaps$CV = 0 * election.leaps$Cp
for (i in 1:nrow(election.leaps$which)) {
    subset = c(1:ncol(X))[election.leaps$which[i,]]
    if (length(subset) > 1) {
        Xw = X[,subset]
        wlm = glm(V ~ Xw)
        election.leaps$CV[i] = cv.glm(model.frame(wlm), wlm, K=5)$delta[1]
    }
    else {
        Xw = X[,subset[1]]
        wlm = glm(V ~ Xw)
        election.leaps$CV[i] = cv.glm(model.frame(wlm), wlm, K=5)$delta[1]
    }
}
```

As mentioned in class, the $C_p$ score is supposed to roughly approximate
leave-one-out cross validation, so we would expect the cross-validation score
(even the 5-fold) and $C_p$ to be correlated.


```python
%%knitr
plot(election.leaps$Cp, election.leaps$CV, pch=23, bg='orange', cex=2)
```

Here is a plot of the estimated 5-fold cross-validation scores
for the models found by leaps using $C_p$.


```python
%%knitr
plot(election.leaps$size, election.leaps$CV, pch=23, bg='orange', cex=2)
```

Of the models selected by leaps using $C_p$ we can compute the 
best one by cross-validation, and compare it to the best one
using $C_p$. Keep in mind, this model can change because there is some
randomness in the $CV$ scores: the randomness introduced by breaking the
data into $K=5$ groups.

```python
%%knitr
best.model.Cp = election.leaps$which[which((election.leaps$Cp == min(election.leaps$Cp))),]
best.model.Cp

best.model.CV = election.leaps$which[which((election.leaps$CV == min(election.leaps$CV))),]
best.model.CV
```
