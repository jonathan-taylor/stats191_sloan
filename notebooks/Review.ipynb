### One and two sample problems

First, we'll look at the one sample of hypothetical blood pressures we saw
in class. The built-in *t.test* function computes the mean and the usual *T* statistic used to test whether the mean is 0 or not. 

Another common task involves finding confidence intervals.

You can do these calculations with the builtin-function called *t.test*. 
For instance, to test whether the mean decrease in the treated group is 0:

```python
%%knitr
bp.table = read.table('http://stats191.stanford.edu/data/pressure_sample.csv', header=TRUE)
mean(bp.table$BP)
sd(bp.table$BP)
T = mean(bp.table$BP) / (sd(bp.table$BP) / sqrt(50))
T
t.test(bp.table$BP)
2*(1-pt(abs(T),49))
mubar = mean(bp.table$BP)
sdbar = sd(bp.table$BP) / sqrt(50)
U = mubar + sdbar * qt(0.975, 49)
L = mubar - sdbar * qt(0.975, 49)
data.frame(L,U)
```

### Density of a *T* random variable

We will be seeing a lot of Student's *T* random variables. This example
plots the density of a T and compares it to a standard Normal.
You can see that the *T* has heavier tails than the Normal.
In the code, you can also see how to use the density functions of R in this
example as well.

```python
%%knitr
xseq <- seq(-3,3,0.01)
plot(xseq, dt(xseq, 10), xlab='s', ylab='Density -- f(s)', type='l', lwd=3, col='red')
lines(xseq, dnorm(xseq), lty=2, lwd=3, col='orange')
abline(v=qt(0.975,10), lty=1, col='red')
abline(v=qt(0.025,10), lty=1, col='red')
abline(v=qnorm(0.975), lty=2, col='orange')
abline(v=qnorm(0.025), lty=2, col='orange')
legend(-1,0.2, c('t, 10 df', 'Normal'), lty=c(1,2), col=c('red', 'orange'))
```

You can also find quantiles, and probabilities
For example, for a 90% confidence interval 
with 10 degrees of freedom. We need to find
the 95% quantile of a T with 10 degrees of freedom

Why 95%? 0.9=$1-\alpha$, so $\alpha=0.1$, and we need the
$1-\alpha/2=1-0.1/2=1-0.05=0.95$ quantile

```python
%%knitr
qt(0.95,10)
```

    is 0.95

### Simulated confidence interval

This example illustrates the principle of confidence intervals.
I will simulate several draws from the Normal distribution
with mean 0, and construct the usual confidence interval for the mean of
each of the samples. Setting $\alpha=0.15$ below should yield about
500*0.15=75 intervals that do not cover 0, and about
500*0.85=425 intervals that do cover 0.

```python
%%knitr

nsample = 500 # how many samples to generate
nobs = 10  # how many observations in each sample
alpha = 0.15

CI.data = matrix(0, nrow=nsample, ncol=2) # a matrix to store the data

cover = numeric(nsample) # a counter to see how many CIs contain 0
for (i in 1:nsample) {
  CI.data[i,] = t.test(rnorm(nobs), conf.level=1-alpha)$conf.int
  cover[i] = (CI.data[i,1] < 0) * (CI.data[i,2] > 0) # add 1 if
					                                 # CI contains 0
}
sum(cover)/nsample # coverage percentage, should be approx 1-alpha
```

This next bit of code plots some of the
confidence intervals and colors
them red if they cover 0, blue otherwise. Since *alpha=0.15* we should
expect about 34 of the 40 intervals to cover 0, but the number that
actually do is 
random (~ Binomial(40,0.85)).

```python
%%knitr

simulate = function() {
nplot = 40 # how many intervals to plot

plot(c(-2,2), c(1, nplot), type='n', xlab='Confidence Intervals', ylab='Sample')
for (i in 1:min(nsample, nplot)) {
    if (cover[i]) {
      lines(CI.data[i,], rep(i,2), col='red', lwd=2) # add a red bar for
					  # each CI that covers
    }
    else {
      lines(CI.data[i,], rep(i,2), col='blue', lwd=2) # add a blue bar for
					  # each CI that doesn't cover
    }    
  }
  }
simulate()
```
