### Wage vs. education example


In this example, we'll look at the output of *lm* for the wage
data and verify that the equations we had for the 
least squares solutions agree with the output.

The data was compiled from a study in econometrics [Learning about Heterogeneity in Returns to Schooling]( http://www.econ.queensu.ca/jae/2004-v19.7/koop-tobias/readme.kt.txt).

```python
%load_ext course_modules.knitr_extension
```

```python
%%knitr

url = 'http://stats191.stanford.edu/data/wage.csv'
wages <- read.table(url, sep=',', header=T)

# Tell R to put logwage and education in R's toplevel namespace
attach(wages)

# Fit simple linear regression model
wages.lm = lm(logwage ~ education)

# Plot the data
plot(education, logwage, pch=23, bg='red', cex=2)

# Add a line to the plot indicating the best
abline(wages.lm, lwd=3, col='yellow')
```

Compute coefficients by hand to make sure they agree

```python
%%knitr
beta.1.hat <- cov(education, logwage) / var(education)
beta.0.hat <- mean(logwage) - beta.1.hat * mean(education)
c(beta.1.hat, beta.0.hat)
```

Estimate sigma squared -- df.resid are degrees of freedom

```python
%%knitr
sigma.hat <- sqrt(sum(resid(wages.lm)^2) / wages.lm$df.resid)
sigma.hat
```

Can you find where $\hat{\beta}_0$, $\hat{\beta}_1$ and
$\hat{\sigma}$ appear?

```python
%%knitr
summary(wages.lm)
```

### Loss functions

We use least squares to find the slope, but we could use
other loss functions. This example plots the loss 
as a function of two parameters. Note that the
least squares one is smoother (hence we would expect it to be easier
to minimize) then the absolute deviation.

```python
%%knitr

X = rnorm(50)
Y = 2 + 1.5 * rnorm(50)

# Squared error loss: 'least squares'

squared.error = function(b0, b1) {
   return(sqrt(sum((Y - b0 - b1*X)^2)))
   }

# Absolute value loss: 'absolute deviation'

absolute.loss = function(b0, b1) {
   return(sum(abs(Y - b0 - b1*X)))
}

# Plot the loss over a grid of values

plot.loss = function(loss, b0, b1) {
   l = matrix(0, length(b0), length(b1))
   for (i in 1:length(b0)) {
         for (j in 1:length(b1)) {
             l[i,j] = loss(b0[i], b1[j])
         }
    }
    image(b0, b1, l, col=rainbow(1000))
}
```

Here is the plot for squared error loss

```python
%%knitr
plot.loss(squared.error, seq(0,4,length=100), seq(-2,2,length=100))
```

And the absolute value loss.

```python
%%knitr
plot.loss(absolute.loss, seq(0,4,length=100), seq(-2,2,length=100))
```

### Confidence and prediction intervals from linear models

R has many builtin convenience functions to compute confidence intervals
for you, the main one being the function *predict*. 

We can easily get confidence intervals for the coefficients
in the regression model using *confint*. These are
90% confidence intervals for the intercept, *(Intercept)*
and the slope, *education* of the true regression line.

```python
%%knitr
confint(wages.lm, level=0.90)
```

We might want a 90% confidence interval for the average logwage
of someone with 14 years education.

```python
%%knitr
predict(wages.lm, list(education=14), interval='confidence', level=0.90)
```

Or, a prediction interval. Note that the prediction interval is wider
because it includes the population variability.

```python
%%knitr
predict(wages.lm, list(education=14), interval='prediction', level=0.90)
detach(wages)
```

### Indicator variables

In this example, we show that the usual two-sample *t*-test
can be viewed as a regression.

```python
%%knitr
X = c(4,5,2,8,3.4,1)
Y = c(3,7,5)

# Join the vectors together

Z = c(X, Y)
dummy = c(1,1,1,1,1,1,0,0,0)

summary(lm(Z ~ dummy))
confint(lm(Z ~ dummy))
t.test(X,Y,var.equal=T)
```

In R, you can create dummy variables using what are known as factors.

```python
%%knitr
dummy = factor(c(rep('A',6), rep('B',3)))
dummy
summary(lm(Z ~ dummy))
confint(lm(Z ~ dummy))
```
