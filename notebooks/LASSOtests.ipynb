```python
import regreg.api as rr, numpy as np

n, p, j = 500, 40, 3
mu, mu_slice = 3, slice(0,10)
X = np.random.standard_normal((n,p))

beta = np.zeros(p)
beta[mu_slice] = mu
Y = np.random.standard_normal(n) + np.dot(X, beta)

lmax = np.fabs(np.dot(X.T,Y)).max()
```

```python
def solve(X, Y, lagrange_values, j):
    """
    Solve the usual LASSO and return:

    T:  Minimized objective value, $\lambda$ = lagrange
    Tj: Minimized objective value under constraint $\beta_j=0, \lambda=$ lagrange
    T0: Minimized objective value, $\lambda=0$
    T0j: Minimized objective value under constraint $\beta_j=0, \lambda=0$
    """
    n, p = X.shape
    
    if lagrange_values[0] != 0:
        lagrange =  [0] + list(lagrange_values)
    return_values = []
    
    # Find the restricted design matrix 
    
    include = np.ones(p, np.bool)
    include[j] = False
    X_slice = X[:,include]

    lagrange = lagrange_values[-1]
    
    # Compute T
    pen = rr.l1norm(p,lagrange=lagrange)
    loss = rr.squared_error(X,Y)
    problem = rr.simple_problem(loss, pen)
    
    penj = rr.l1norm(p-1, lagrange=lagrange)
    lossj = rr.squared_error(X_slice, Y)
    problemj = rr.simple_problem(lossj, penj)
    
    for lagrange in lagrange_values[::-1]:
        pen.lagrange = lagrange
        penj.lagrange = lagrange
        
        coef = problem.solve(tol=1.e-10)
        T = problem.objective(coef)
        
        coefj = problemj.solve(tol=1.e-10)
        Tj = problemj.objective(coefj)
        
        return_values.append(Tj-T)
        
    return np.array(lagrange_values[::-1]), np.array(return_values)
    
```

Let's look at the test statistic for an active coefficient and a sequence of $\lambda$:

```python
lagrange_values = np.linspace(0,10,101)
lagrange_values, path_of_test_statistics = solve(X, Y, lagrange_values, 3)
```

```python
import matplotlib.pyplot as plt
plt.plot(lagrange_values, path_of_test_statistics, label='Proposed test statistic', linewidth=2)
plt.plot([lagrange_values[0], lagrange_values[-1]], [path_of_test_statistics[-1]]*2, '--', label='Upper bound?', linewidth=3)
plt.legend(loc = 'lower right')
```

Now let's try it for an inactive coefficient.

```python
lagrange_values = np.linspace(0,10,101)
lagrange_values, path_of_test_statistics = solve(X, Y, lagrange_values, 20)
```

### Another bound?

Another bound we can use is that the above test statistic is bounded by the difference in objective when fitting
a LASSO with $X_j$ unpenalized and when $X_j=0$.

```python
def solve2(X, Y, lagrange_values, j):
    """
    Solve the usual LASSO and return:

    T:  Minimized objective value, $\lambda$ = lagrange
    Tj: Minimized objective value under constraint $\beta_j=0, \lambda=$ lagrange
    T0: Minimized objective value, $\lambda=0$
    T0j: Minimized objective value under constraint $\beta_j=0, \lambda=0$
    """
    n, p = X.shape
    
    if lagrange_values[0] != 0:
        lagrange =  [0] + list(lagrange_values)
    return_values = []
    
    # Find the restricted design matrix 
    
    include = np.ones(p, np.bool)
    include[j] = False
    X_slice = X[:,include]

    lagrange = lagrange_values[-1]
    
    # Compute T
    pen = rr.l1norm(p,lagrange=lagrange)
    loss = rr.squared_error(X,Y)
    problem = rr.simple_problem(loss, pen)
    
    penj = rr.l1norm(p-1, lagrange=lagrange)
    lossj = rr.squared_error(X_slice, Y)
    problemj = rr.simple_problem(lossj, penj)
    
    selector = rr.selector(include, (p,))
    
    problem_unpen = rr.separable_problem(loss, (p,), [penj], [include])
    
    for lagrange in lagrange_values[::-1]:
        pen.lagrange = lagrange
        penj.lagrange = lagrange
        penj_unpen = lagrange
        
        coef = problem.solve(tol=1.e-10)
        T = problem.objective(coef)
        
        coefj = problemj.solve(tol=1.e-10)
        Tj = problemj.objective(coefj)
        
        coef_unpen = problem_unpen.solve(tol=1.e-10)
        T_unpen = problem_unpen.objective(coef_unpen)
        
        return_values.append((Tj-T, Tj-T_unpen))
        
    return np.array(lagrange_values[::-1]), np.array(return_values)
    
```

```python
lagrange_values = np.linspace(0,10,21)
lagrange_values, path_of_test_statistics = solve2(Xnew, Ynew, lagrange_values, 20)
```

```python
plt.plot(lagrange_values, path_of_test_statistics[:,0], label='Proposed test statistic', linewidth=2)
plt.plot(lagrange_values, path_of_test_statistics[:,1], label='Upper bound?', linewidth=2)
plt.legend(loc = 'upper right')
```
