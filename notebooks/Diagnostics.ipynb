With the package installed. You can load it with the *library* function.


```python
%load_ext course_modules.knitr_extension
```

```python
%%knitr
library(car)
```

```python
%%knitr
url = 'http://stats191.stanford.edu/data/scottish_races.table'
races.table = read.table(url, header=T)
attach(races.table)
```

Now, we plot the data to be used in our model. As expected, *Time*
does increase with *Distance* and *Climb*. We also fit a model to the data.


```python
%%knitr
plot(races.table[,3:5], pch=23, bg='orange', cex=2)
races.lm = lm(Time ~ Distance + Climb)
```

### Standard diagnostic plots

```python
%%knitr
par(mfrow=c(2,2))
plot(races.lm, pch=23 ,bg='orange',cex=2)
```

### DFBETAS

```python
%%knitr
plot(dfbetas(races.lm)[,'Climb'], pch=23, bg='orange', cex=2, ylab="DFBETA (Climb)")
plot(dfbetas(races.lm)[,'Distance'], pch=23, bg='orange', cex=2, ylab="DFBETA (Distance)")
```

### Cook's distance

```python
%%knitr
plot(cooks.distance(races.lm), pch=23, bg='orange', cex=2, ylab="Cook's distance")
```

### DFFITS

```python
%%knitr
plot(dffits(races.lm), pch=23, bg='orange', cex=2, ylab="DFFITS")
```

### Bonferroni outlier test

```python
%%knitr
outlierTest(races.lm)
```

### All influence measures

```python
%%knitr
influence.measures(races.lm)
```

### Added variable plots

```python
%%knitr
avPlots(races.lm, "Climb")
avPlots(races.lm, "Distance")
```

### Component + residual plots

```python
%%knitr
crPlots(races.lm, 'Climb')
crPlots(races.lm, 'Distance')
```

### Hat values

```python
%%knitr
plot(hatvalues(races.lm), pch=23, bg='orange', cex=2, ylab='Hat values')
```
