The sums of squares and $R^2$ are defined analogously
to those in simple linear regression.


```python
%%knitr
SST = sum((Y - mean(Y))^2)
MST = SST / (length(Y) - 1)
SSE = sum(resid(supervisor.lm)^2)
MSE = SSE / supervisor.lm$df.residual
```

The adjusted $R^2$ is expressed in terms
of mean squares rather
than sums of squares

```python
%%knitr
R2 = 1 - SSE/SST
R2.adj = 1 - MSE / MST

data.frame(R2,R2.adj)
```

### Partial regression coefficients


The estimated coefficients in a multiple linear regression
model are known as *partial regression coefficients*.
This is because
the coefficient for, say, *X6* can be thought
of as the simple linear regression coefficient
of *Y* on *X6* after regressing out, or controlling for, everything
else. 
More formally,  the coefficient
for *X6* is what you would get if you first regress *Y* and *X6* onto 
everything except *X6*, and take
the residuals, creating two new variables, *Z* and *W*
and then fit a simple linear regression model with *Z* as the outcome
and *W* as the covariate.


```python
%%knitr
supervisor.lm$coef
```

Regress X6 onto everything else
and call the residuals W
this is the part of X6 that is 
not explained by any of the other variables.

```python
%%knitr
W = resid(lm(X6 ~ X1 + X2 + X3 + X4 + X5, supervisor.table))
```

Do the same for Y.
  

```python
%%knitr
Z = resid(lm(Y ~ X1 + X2 + X3 + X4 + X5, supervisor.table))
```

Note that coefficient of W is the same as in original model

```python
%%knitr
print(lm(Z~W)$coef["W"])
```

### Matrix representation of coefficients

We can construct the least squares estimates by hand, 
using matrices.

```python
%%knitr
n = length(Y)
X = cbind(rep(1,n), X1, X2, X3, X4, X5, X6)
colnames(X)[1] = '(Intercept)'
X[1:5,]
```

The matrix X is the same as formed by R

```python
%%knitr
model.matrix(supervisor.lm)[1:5,]
```

Solve for $\hat{\beta}$

```python
%%knitr
beta = solve(t(X) %*% X) %*% t(X) %*% Y
beta
coef(supervisor.lm)
```

The fitted values are $\hat{Y}=X\hat{\beta}$:

```python
%%knitr
Y.hat = X %*% beta
```

Now, let's compute the covariance matrix:


```python
%%knitr
sigma.hat = sqrt(sum((Y - Y.hat)^2) / (n - ncol(X)))
cov.beta = sigma.hat^2 * solve(t(X) %*% X)
cov.beta
```

We've seen the diagonal of this matrix before, its square root
is the Std. Error column of summary(supervisor.lm)

```python
%%knitr
sqrt(diag(cov.beta))
```

There's an easier way to find the covariance matrix

```python
%%knitr
vcov(supervisor.lm)
```

### Confidence intervals by hand

In this example, we write a function that
explicitly computes a confidence interval
(and can be used for predicition intervals 
with the "extra" argument).
This exercise shows the calculations that R 
is doing under the hood: the function *predict*
is generally going to be fine for our
purposes.

```python
%%knitr
CI.lm = function(cur.lm, a, level=0.95, extra=0) {

     # the center of the confidence interval
     center = sum(a*cur.lm$coef)

     # the estimate of sigma^2
     sigma.sq = sum(resid(cur.lm)^2) / cur.lm$df.resid

     # the standard error of sum(a*cur.lm$coef)
     se = sqrt(extra * sigma.sq + sum((a %*% vcov(cur.lm)) * a))

     # the degrees of freedom for the t-statistic
     df = cur.lm$df
     
     # the quantile used in the confidence interval

     q = qt((1 - level)/2, df, lower.tail=F)

     # upper, lower limits
     U = center + se * q
     L = center - se * q
     return(data.frame(center, L, U))
}
```

With this function, we can do basically anything
that *predict* did, as far as confidence intervals go.

```python
%%knitr
CI.lm(supervisor.lm, c(1,65,50,55,64,75,40))
```

```python
%%knitr
predict(supervisor.lm,list(X1=65,X2=50,X3=55,X4=64,X5=75,X6=40),interval='confidence')
```

By using the *extra* argument, we can make
prediction intervals.

```python
%%knitr
CI.lm(supervisor.lm, c(1,65,50,55,64,75,40), extra=1)
```

```python
%%knitr
predict(supervisor.lm,list(X1=65,X2=50,X3=55,X4=64,X5=75,X6=40),interval='prediction')
```

If we want, we can set the intercept term to 0. This allows
us to construct confidence interval for, say, 
how much a supervisor's rating will increase
if we change *X1* and *X2* each by 5 points, leaving
everything else unchanged. Therefore, what we want is a 
confidence interval for
5 times the coefficient of X1 + 5 times the coefficient
of X2.

```python
%%knitr
CI.lm(supervisor.lm, c(0,5,5,0,0,0,0))
```

Most of the time, *predict* will do what you want so this 
won't be used too often.

### A general *F* test

When comparing two models, one a special case of the other (i.e. 
one nested in the other), we can test if the smaller
model (the special case) is roughly as good as the 
larger model in describing our outcome. This is typically
tested using an *F* test based on comparing
the two models. The following function does this.
 

```python
%%knitr

f.test.lm = function(R.lm, F.lm) {
    SSE.R = sum(resid(R.lm)^2)
    SSE.F = sum(resid(F.lm)^2)
    df.num = R.lm$df - F.lm$df
    df.den = F.lm$df
    F = ((SSE.R - SSE.F) / df.num) / (SSE.F / df.den)
    p.value = 1 - pf(F, df.num, df.den)
    return(data.frame(F, df.num, df.den, p.value))
}
```

We can try dropping variables *X2* and *X3* with our new function
and see the results as an *F* test.

```python
%%knitr
reduced.lm = lm(Y ~ X1 + X4 + X5 + X6)
f.test.lm(reduced.lm, supervisor.lm)
anova(reduced.lm, supervisor.lm)
```

This method of comparing to models works in *almost* all
examples we'll see this quarter.

### Imposing constraints on parameters

For instance, we might suppose that the
coefficients for *X1* and *X3* are the same
and want to test this. We do this, again, by
comparing a "full model" and a "reduced model".

For simplicity, we will just consider a model with only variables 
*X1* and *X3* and test whether the coefficients of *X1* and *X3* are equal.


```python
%%knitr
reduced.lm = lm(Y~X1+X3)
```

As discussed in the notes, we can do this simply by
making a new regressor 

```python
%%knitr
Z = X1 + X3
equal.lm = lm(Y ~ Z)
print(anova(equal.lm, reduced.lm))
```

The book also considers an example $\beta_1+\beta_3=1$.
This time, since the outcomes are different, we can't use
*anova*, but *f.test.lm* does work

```python
%%knitr
Z = X1 - X3
W = Y - X3
print(anova(lm(W ~ Z), reduced.lm))
print(f.test.lm(lm(W ~ Z), reduced.lm))
```

What we had to do above was subtract *X3* from *Y* on the right hand
side of the formula. R has a way to do this called using an *offset*.

```python
%%knitr
print(anova(lm(Y ~ Z, offset=X3), reduced.lm))
```

Looking at the resulting models, they are almost identical, the 
$R^2$ is different because in the first model, the
outcome is *Y*, and the second, the outcome is *Y-X3*.

```python
%%knitr
print(summary(lm(Y ~ Z, offset=X3)))
print(summary(lm(W ~ Z)))
detach(supervisor.table)
```
