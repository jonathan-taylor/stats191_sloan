### Logistic transform

The logistic curve is a sigmoidal curve starting at 0 at $-\infty$, and 
rising to 1. Hence, it is a distribution function, also called the 
logistic distribution.
The logistic transform, or *logit* is the quantile
function of this distribution.


```python
%load_ext course_modules.knitr_extension
```

```python
%%knitr
g <- function(x) {
    return(log(x / (1 - x)))
}

g.inv <- function(y) {
    return(exp(y) / (1 + exp(y)))
}

x = seq(0.01,0.99,length=200)
plot(x, g(x), lwd=2, type='l', col='red')
    
p = seq(g(0.01), g(0.99), length=200)
plot(p, g.inv(p), lwd=2, type='l', col='red')
```

### Flu shot example

This is the flu data discussed in the notes. We will fit a logistic
regression model to it using the *R* function *glm*.

```python
%%knitr
url = 'http://stats191.stanford.edu/data/flu.table'
flu.table = read.table(url, header=T)
pairs(flu.table, cex.labels=3, pch=23, bg='orange', cex=2)
```

Fitting the model is very much like using *lm* only we have
to specify what is called the *family*. 

```python
%%knitr
flu.glm = glm(Shot ~ Age + Health.Aware, data=flu.table, family=binomial())
summary(flu.glm)
```

Diagnostics can be plotted just as for *lm*. The residuals
used here are what are called *deviance residuals*, not just
$Y_i-\hat{Y_i}$.

```python
%%knitr
par(mfrow=c(2,2))
plot(flu.glm)
```

Influence measures can also be investigated for *glm*s.

```python
%%knitr
library(car)
influence.measures(flu.glm)
```

The function *predict* works just as for *lm* but it does not 
always give you output on the scale of the *response*, which
is 0-1 for binary responses.


```python
%%knitr
logodds = predict(flu.glm, list(Age=c(35,45),Health.Aware=c(50,50)))
logodds
exp(logodds)/(1+exp(logodds))
```

To get a direct value on a binary scale, use the *type* argument

```python
%%knitr
pi.hat = predict(flu.glm, list(Age=c(35,45),Health.Aware=c(50,50)), type="response")
pi.hat
```

In class, we discussed the odds ratios and their interpretations as
approximately ratios of probabilities. This is only
accurate when the outcome is rare, known as the *rare disease* hypothesis.


```python
%%knitr
exp(logodds[2]-logodds[1])
pi.hat[2]/pi.hat[1]
```

### Comparing models

Comparing models is done with deviance instead of sums of squares. The
function *anova* will do this automatically for us.
Note that the *p*-value is different using *anova* and the
*summary*. This is because the *p*-value in the summary table
is based on a Normal approximation and not a difference of deviances.
In the *lm* case, these two things are identical, but not 
necessarily in the *glm* case. The test using the
summary table is called a *Wald test*, whereas the deviance
one is called a *likelihood ratio test*.


```python
%%knitr
reduced.glm = glm(Shot ~ Health.Aware, family=binomial(), data=flu.table)
anova(reduced.glm, flu.glm)
pval = 1-pchisq(16.863,1)
pval
summary(flu.glm)

 
```

Confidence intervals are also slightly different.
They are constructed using the likelihood, 
not the summary table (which would be *Wald* confidence intervals).


```python
%%knitr
confint(flu.glm)
```

The Wald interval for *Age* is

```python
%%knitr
center = coef(flu.glm)['Age']
SE = sqrt(vcov(flu.glm)['Age', 'Age'])
U = center + SE * qnorm(0.975)
L = center + SE * qnorm(0.025)
data.frame(center, L, U)
```

Model selection is straightforward.

```python
%%knitr
step.glm = step(flu.glm, scope=list(upper=~.^2), direction='both')
summary(step.glm)
```

We also talked about *probit* and *cloglog* models in class. This is specified using the *link* argument to *binomial*, which defaults to *logit*.

```python
%%knitr
summary(glm(Shot ~ Age + Health.Aware, data=flu.table, family=binomial(link='probit')))
summary(glm(Shot ~ Age + Health.Aware, data=flu.table, family=binomial(link='cloglog')))
```

### IRLS (Iteratively Reweighted Least Squares)

Logistic, or binary regression models, can be fit using
Iteratively Reweighted Least Squares (IRLS). This 
example below shows how this might
work. For logistic regression, this algorithm
is a direct implementation of the Newton-Raphson algorithm
used to minimize smooth functions.
In *R*, all of this is done for you by *glm*.


```python
%%knitr

g.prime = function(x) {
    return(1/(x*(1-x)))
}

V = function(mu) {
    return(mu*(1-mu))
}

niter = 10
mu = rep(mean(flu.table$Shot), length(flu.table$Shot))

for (i in 1:niter) {
    Z = g(mu) + g.prime(mu) * (flu.table$Shot - mu)
    W = 1 / (g.prime(mu)^2 * V(mu))
    Z.lm = lm(Z ~ Age + Health.Aware, weights=W, data=flu.table)
    eta = predict(Z.lm)
    mu = g.inv(eta)
    beta = Z.lm$coef
}
```

Let's compare to the results from *glm*

```python
%%knitr
X = model.matrix(Z.lm)
SE.beta = sqrt(diag(solve(t(X) %*% diag(W) %*% X)))
T.beta = beta / SE.beta

cbind(beta, SE.beta, T.beta)
summary(flu.glm)$coef
```
